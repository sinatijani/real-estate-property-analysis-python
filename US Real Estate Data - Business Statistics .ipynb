This step may be an over-stretch for others. But I am trying to prune the dataset to the best rationale state possible. 

By default, duplicated uses Keep=First which keeps first occurrence and shows the rest as duplicate. But by doing the above, it is seen that price, bed, bath, acre_lot, and house_size of the same full_address may differ. 

Assumptions are that, just as there were over 87% of the dataset were duplicates, the real estate dataset was gathered from different agencies and  sources which led to duplicates of properties. Also, data about the same property were gathered at differet states hence difference in beds, baths, house size designed and allotted for them.

Other observation is, the duplicate with the highest prices has higher number of bed, bath, and/or house_size. 

On this basis, I would assume that the duplicate with the highest price and house_size represents the final state of the duplicate to be sold and therefore, their entry with the highest prices should be kept and others dropped.

Let's start with City since it has the least missing values.

```python
# finding missing cities

missing_cities = df5[df5.city.isnull()]
missing_cities
```

Inspecting if their full addresses could reveal their cities doesn't work. Some libraries like usaddress or APIs like Google Geocoding API or Nominatim (OpenStreetMap) could be tried to identify the cties but it is not worth it for this dataset as there are only 21 of them. I have made the decision to drop them fromt he dataset.

```python
# removing rows with empty cities

df5.dropna(subset='city', inplace=True)
```

```python
# confirming the emty cities are removed

df5.isnull().sum()
```

Filling the missing bed and bath with their mean or median would not be good as housing prices have irregular correlation with no. of beds and baths as we found in the scatter plots. The location determines the prices therefore it would make more sense to fill the empty values with mean/median values from their respective states or cities. Using the cities would reduce margin of error compared to state and compared to the whole dataset.

```python
# calculating the medians of each city

city_medians = df5.groupby('city')[['bed', 'bath', 'acre_lot', 'house_size']].transform('median')
city_medians
```

```python
# filling the missing bed, bath, acre_lot and house_size values in each city with it's median

df5['bed'] = df5['bed'].fillna(city_medians['bed'])
df5['bath'] = df5['bath'].fillna(city_medians['bath'])
df5['acre_lot'] = df5['acre_lot'].fillna(city_medians['acre_lot'])
df5['house_size'] = df5['house_size'].fillna(city_medians['house_size'])
```

```python
# confirming that the missing values are filled

df5.isnull().sum()
```

As stated earlier, the missing dates are left intentionally as they may signify properties that are not yet sold.
